### **1.3 Analytic Approach:**

To address the problem, we can adopt a regression modeling approach. We will utilize the provided dataset, "data_daegu_apartment.csv", containing information about apartments in Daegu, to train and evaluate a regression model. The model will learn the relationships between the features (e.g., hallway type, proximity to subway stations, number of facilities nearby, apartment size) and the target variable (sale price).

The steps involved in the analytic approach are as follows:

**1. Data preprocessing:** Perform data cleaning, handle missing values if any, and encode categorical variables.

**2. Exploratory data analysis:** Analyze the distribution of the target variable and explore the relationships between the features and the sale price.

**3. Feature selection:** Identify the most relevant features that strongly influence the sale price using techniques such as correlation analysis and feature importance.

**4. Model selection:** Choose an appropriate regression model (e.g., linear regression, decision tree regression, random forest regression, etc.) to train on the dataset.

**5. Model training and evaluation:** Split the dataset into training and testing sets, train the selected model on the training data, and evaluate its performance using appropriate regression evaluation metrics (e.g., mean squared error, mean absolute error).

**6. Model refinement and optimization:** Fine-tune the model hyperparameters, consider feature engineering techniques, and iteratively improve the model's performance.

**7. Final model deployment:** Once satisfied with the model's performance, deploy the model to make predictions on new, unseen data.

### **1.4 Metric Evaluation:**

The following evaluation metrics can be used to assess the performance of the regression model:

**1. Mean Squared Error (MSE):** It measures the average squared difference between the predicted sale prices and the actual sale prices. A lower MSE indicates better model performance.

**2. Root Mean Squared Error (RMSE):** It is the square root of the MSE and provides a measure of the average absolute deviation of the predicted sale prices from the actual sale prices. Similar to MSE, a lower RMSE indicates better accuracy.

**3. Mean Absolute Error (MAE):** It calculates the average absolute difference between the predicted sale prices and the actual sale prices. Similar to MSE and RMSE, a lower MAE indicates better accuracy.

**4. Mean Absolute Percentage Error (MAPE):** It measures the average percentage difference between the predicted sale prices and the actual sale prices. MAPE provides a relative measure of the prediction error. A lower MAPE indicates better accuracy.

These metrics will help assess the accuracy and reliability of the regression model in predicting the sale prices of apartments in Daegu.

#### **2.3.2 Correlation Matrix**

A correlation matrix is a useful tool that measures the strength and direction of relationships between variables in a dataset. It provides a numerical value, the correlation coefficient, which ranges from -1 to 1. Positive correlations indicate that variables tend to increase or decrease together, while negative correlations mean that one variable increases as the other decreases. The correlation matrix helps us identify redundant variables, prioritize influential ones, and gain insights into the interdependencies among variables. It is a valuable tool for exploring and understanding relationships within a dataset.

```python
plt.figure(figsize = (8,8))
df_corr = df.corr(method='spearman')
cmap = sns.color_palette('Blues', as_cmap=True)
ax = sns.heatmap(data=df_corr, cmap=cmap, annot = True)

# Rotate x-axis tick labels by 45 degrees
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.title('Correlation Matrix', fontsize=16)

plt.show()
```

```python
df.corr().round(2)
```

**Insights:**

1. "SalePrice" has a negative correlation with "N_FacilitiesNearBy(ETC)", "N_FacilitiesNearBy(PublicOffice)", "N_SchoolNearBy(University)", "YearBuilt", "N_FacilitiesInApt", and "Size(sqf)". This suggests that as these variables increase, the "SalePrice" tends to decrease.

2. "SalePrice" has a positive correlation with "N_Parkinglot(Basement)" and "Size(sqf)". This implies that as the number of parking lots in the basement and the size of the property increase, the "SalePrice" tends to increase as well.

3. Among the variables, "Size(sqf)" shows the highest positive correlation with "SalePrice" (correlation coefficient of 0.70), indicating that the size of the property has a relatively strong influence on its sale price.

**Encoding**

```python
# Mapping for TimeToSubway
ordinal_mapping = [
    {'col':'TimeToSubway',
     'mapping': {'0-5min':1, '5min~10min':2, '10min~15min':3, '15min~20min':4, 'no_bus_stop_nearby':5}}
]

# Pipeline for TimeToSubway
pipe_ordinal_robust = Pipeline([
    ('encoder', OrdinalEncoder(mapping = ordinal_mapping)),
    ('scaler', RobustScaler()),
])

pipe_ordinal_robust
```

The ColumnTransformer in scikit-learn allows for the selection of specific columns or subsets of columns in a dataset and applies different preprocessing steps to each selected subset. It facilitates parallel execution of these transformations, improving the efficiency of the preprocessing pipeline. The ColumnTransformer seamlessly integrates with scikit-learn's Pipeline functionality and enables the creation of end-to-end data preprocessing and modeling pipelines. Additionally, we use a one-hot encoder for the "HallwayType" column to create separate columns for each type of hallway. Finally, we apply a robust scaler to all the remaining numerical features in our dataset. These transformations help make the data suitable for machine learning algorithms.

### **4.1. Cross Validation**

Cross-validation is a technique in machine learning that is used to assess the performance of a regression model. It helps to estimate how well the model generalizes to unseen data by evaluating its performance on multiple subsets of the training data.

![](https://i0.wp.com/sqlrelease.com/wp-content/uploads/2021/07/K-fold-cross-validation-1.jpg?ssl=1)


In 5-fold cross-validation, the trainset is split into five equal-sized subsets. The model is trained and evaluated five times, using a different subset as the validation set each time. This ensures that each subset serves as the validation set exactly once. Performance metrics, like mean squared error or R-squared, are calculated for each iteration, and the average performance across all folds is used to estimate the model's overall performance. This technique helps prevent overfitting or underfitting and provides a more reliable assessment of the model's predictive ability on unseen data by utilizing multiple validation sets.

#### **4.1.1 Cross Validation Using Linear Regression**

In linear regression, it is important to verify certain assumptions for the model to be valid and reliable. Three key assumptions include checking for multicollinearity, assessing the normality of residuals (errors), and ensuring the Gauss-Markov assumption holds. Multicollinearity refers to the presence of high correlation between predictor variables, which can affect the model's stability and interpretation. To assess this, one method is to calculate the Variance Inflation Factor (VIF), which measures the extent of correlation between predictors. Additionally, examining the normality of residuals helps ensure that the errors are normally distributed, allowing for accurate statistical inference. Lastly, the Gauss-Markov assumption states that the errors should have constant variance and be uncorrelated. By evaluating these assumptions, we can gauge the reliability of the linear regression model and confidently interpret the results.

The negative signs were included because scikit-learn uses negation for certain scoring metrics. 

In scikit-learn, some scoring metrics are defined such that higher values are better, while others are defined such that lower values are better. To maintain consistency, scikit-learn represents the scoring metrics as negative values when lower values are better. This allows all metrics to be maximized during model evaluation.

For example, in the case of mean squared error (MSE), a lower value indicates better performance. However, scikit-learn's cross-validation functions expect scoring metrics to be maximized, so MSE is multiplied by -1 to convert it to a maximization problem. The same applies to other metrics like mean absolute error (MAE) and mean absolute percentage error (MAPE).

Therefore, when retrieving the scores using `cv_score`, we need to negate the values to obtain the actual error values (positive values for MSE, MAE, and MAPE). By negating the values, we revert them to their original form.

### **4.2 Modeling: HyperParameter Tuning using GridSearch**

Since we have found the best algorithm based on the lowest RMSE_mean and RMSE_std, next we run HyperParameter Tuning using GridSearch. Hyperparameter tuning using grid search is a technique that helps find the best combination of settings for a machine learning model. Think of hyperparameters as the settings or knobs of a model that need to be adjusted to achieve optimal performance. Grid search works by systematically trying out different combinations of these settings from a predefined grid or range of values. It then evaluates the model's performance using a specified evaluation metric for each combination of settings. The goal is to find the combination that yields the best performance according to the evaluation metric. It's like exploring a grid of possibilities to find the sweet spot where the model performs the best. By exhaustively searching through the grid of hyperparameters, grid search helps us discover the optimal settings that maximize the model's performance and improve its accuracy, making it a powerful tool for optimizing machine learning models.

```python
from sklearn.model_selection import GridSearchCV    # Package for Hyperparameter Tuning using GridSearch
```

```python
# Package for Hyperparameter Tuning using GridSearch
from sklearn.model_selection import GridSearchCV    

# Create the Gradient Boosting Regressor model
gboost_model = GradientBoostingRegressor()

# Create a pipeline for preprocessing and model training
pipeline = Pipeline([
    ('transformer', transformer),           # Apply the transformer to preprocess the data
    ('model', gboost_model)                 # Train the Gradient Boosting model
])

# Define the hyperparameter grid to search over
hyperparam_space = {
    'model__n_estimators': [50, 100, 200, 300],                      # Number of boosting stages to perform                                          
    'model__learning_rate': [0.1, 0.05, 0.01, 0.001],                # Learning rate for the Gradient Boosting model                                       
    'model__max_depth': [3, 5, 7, 9],                                # Maximum depth of each tree in the Gradient Boosting model                                       
    'model__min_samples_split': [2, 5, 10, 15],                      # Minimum number of samples required to split an internal node                                      
    'model__min_samples_leaf': [1, 2, 4, 6, 8],                      # Minimum number of samples required to be at a leaf node                                      
    'model__subsample': [0.8, 0.9, 1.0],                             # Subsample ratio of the training instances                                
    'model__max_features': ['sqrt', 'log2', None],                   # Maximum number of features to consider for each split  
}

# Perform grid search cross-validation
grid_search = GridSearchCV(
    estimator = pipeline,
    param_grid = hyperparam_space,
    cv=5,                                # 5-fold cross-validation
    scoring='neg_mean_squared_error',    # Evaluation metric: negative mean squared error
    n_jobs= -1)                          # Utilize all available CPU cores for parallel processing
```

![](https://miro.medium.com/v2/resize:fit:788/1*pEu2LNmxf9ttXHIALPcEBw.png)

Boosting is a step-by-step method that improves a model (called a weak learner) by focusing on the mistakes it makes. The goal is to turn the weak learner into a strong one. A weak learner is a model that performs slightly better than random guessing. In bagging and random forests, each tree is built separately, but in boosting, trees are grown one after another and learn from the mistakes of the previous trees.

**Hyperparameters:**

- **n_estimators**: This parameter determines the number of boosting stages or the number of decision trees to be created. Increasing the number of estimators can improve the model's performance but also increase training time.

- **learning_rate**: The learning rate controls the contribution of each tree in the ensemble. A lower learning rate means that each tree's influence is smaller, requiring more estimators to achieve the same level of accuracy.

- **max_depth**: This parameter limits the depth of each tree in the ensemble. Deeper trees can capture more complex relationships but may overfit the data. It's important to find the right balance to avoid overfitting or underfitting.

- **min_samples_split**: This parameter sets the minimum number of samples required to split an internal node. Higher values can prevent overfitting by ensuring a minimum number of samples in each split.

- **min_samples_leaf**: The min_samples_leaf parameter determines the minimum number of samples required to be at a leaf node. Similar to min_samples_split, it helps control overfitting by ensuring a minimum number of samples at the leaves.

- **subsample**: Subsampling refers to randomly selecting a subset of the training instances for each boosting iteration. It helps reduce overfitting and can improve generalization. The subsample parameter determines the fraction of samples to be used for training each tree.

- **max_features**: This parameter controls the maximum number of features to consider when looking for the best split in each tree. The options 'sqrt', 'log2', and None specify different strategies for determining the maximum number of features.

By tuning these parameters through grid search cross-validation, the Gradient Boosting algorithm optimizes the model's performance by finding the best combination of parameters that minimizes the mean squared error. This allows the model to make more accurate predictions by leveraging the strengths of multiple decision trees and effectively capturing complex relationships within the data.
